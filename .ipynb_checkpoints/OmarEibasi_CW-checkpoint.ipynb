# <center>Peer-Reviewed paper<center>

The reviewed paper: https://iopscience.iop.org/article/10.1088/1742-6596/1755/1/012006/pdf
        
In this paper we are going to test face recognition and identification using deep learning approach
Which has been proposed by G. Thomson.

The purpose of the paper is to find other the traditional face recognition models which will solve
some difficulties such as Misalignment, Pose Variation, Illumination Variation and Expression
Variation Multiple app.

To solve this problem, they have used a face recognition and identification system which is designed and developed using deep learning approach with the CNN model to train the data.

Deep learning can achieve a nice approximation of a complex function through increment of hidden
layers, that teaches computer to do as what human does naturally. The system will be tested via
recognition of three peoples with multiple times at different locations, mainly to test how light
intensity affect the accuracy of the system. The accuracy is verified using confusion matrix. The
calculation is based on

((TN + TP) / Total) x 100%

where TN is true negative while TP is true positive.

In the end of the research, they have arrived to a conclusion that with a large number of face
images being trained into a classifier can achieve accuracy of 91.7% in recognizing image and
86.7% in real-time video plus if the face is too close the system cant recognize it and better the
light lead to a better result.

another extra conclusion is that OpenCV is much faster than Matlab but Matlab environment is
simpler, more user-friendly and doesnt suffer from Memory allocation and memory leak.

### Pre-processing using PCA

```python
from sklearn.decomposition import PCA
# create dimenssion reduction
pca = PCA(n_components=150, whiten=True, random_state=42)
```

### Training and testing subsets
Divide into training and testing subsets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(faces.data, faces.target,
                                                random_state=42)
```

 Explore the training and testing subsets

```python
print(f'The size of the Training set is {X_train.shape[0]}')
print(f'The size of the Testing set is {X_test.shape[0]}')
# 62 X 47 pixels = 2914
print(f'The number of columns in the training and testing set is {X_train.shape[1]}')
print(f'Type of X_train {type(X_train)}')
print(f'Type of y_train {type(y_train)}')
```

# <center>SVM<center>
Create SVM model and combine it with the PCA images

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
# create your SVM model with RBF kernel
svc = SVC(kernel='rbf', class_weight='balanced')
# your pipline
model = make_pipeline(pca, svc)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'svc__C': [1, 10, 20, 100],
'svc__gamma': [0.0001, 0.0005, 0.001, 0.0001]}
grid = GridSearchCV(model, param_grid)
```

Fit the model

```python
grid.fit(X_train, y_train)
best_params = grid.best_params_
```

Best Model Fit

```python
# best model we have
model = grid.best_estimator_
y_predicted = model.predict(X_test)
```

### Checking the quality of our model(Classification report)
Precision  What percent of your predictions were correct<br>
Recall  What percent of the positive cases did you catch. <br>
F1 score  What percent of positive predictions were correct.<br>
Support  the number of actual occurrences of the class in the specified dataset

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predicted, target_names=faces.target_names))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
# compare the actual label against the predicted label
mat = confusion_matrix(y_test, y_predicted)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,cmap="YlGnBu",
            xticklabels=faces.target_names,
            yticklabels=faces.target_names)
plt.xlabel('Actual Label')
plt.ylabel('Predicted Label');
```

### Save SVM

```python
import pickle
# Save to file in the current working directory
pkl_filename = "model/svm_model.pkl"
# save your model that was created above (lg_model)
with open(pkl_filename, 'wb') as file:
    pickle.dump(model, file)
```

### Test score

```python
# Load from file
with open(pkl_filename, 'rb') as file:
    pickle_model = pickle.load(file)
# Lets test the mode loaded from a file and check results
score = pickle_model.score(X_test, y_test)
print("Test score: {0:.2f} %".format(100 * score))
y_hat = pickle_model.predict(X_test)
```

### save your test set as a data frame

```python
# you can save your test set as a data frame as below
df = pd.DataFrame(X_test)
# add the label
df['Label']=y_test
# save to csv
df.to_csv('test/Xy_test.csv', index=False)
```
